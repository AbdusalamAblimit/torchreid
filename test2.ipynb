```python
import pickle
import cv2
import matplotlib.pyplot as plt

# 替换为你自己的文件路径
pkl_file_path = 'reid-data/market1501/Market-1501-v15.09.15/pose/bounding_box_train/1495_c6s3_085917_09_pose.pkl'  # 或 image_name_float.pkl
img_path = 'reid-data/market1501/Market-1501-v15.09.15/bounding_box_train/1495_c6s3_085917_09.jpg'

# 加载并输出数据尺寸
with open(pkl_file_path, 'rb') as f:
    data = pickle.load(f)

print(f"类型: {type(data)}")
if hasattr(data, 'shape'):
    print(f"数据尺寸: {data.shape}")
elif isinstance(data, list):
    print(f"列表长度: {len(data)}")
    if len(data) > 0 and hasattr(data[0], 'shape'):
        print(f"第一个元素的尺寸: {data[0].shape}")
else:
    print("数据不是数组或列表，无法显示尺寸。")
print(data)

# 读取图像
img = cv2.imread(img_path)
if img is None:
    print("图像读取失败，请检查文件路径。")
else:
    # 将 BGR 转换为 RGB，以便使用 matplotlib 正确显示颜色
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # 创建图像显示窗口
    plt.figure(figsize=(8, 8))
    plt.imshow(img_rgb)
    
    # 绘制 17 个关键点（假设 data 中每个元素为 (x, y) 或 (x, y, conf)）
    for point in data:
        # 如果有三个元素，默认第 3 个为置信度，这里只绘制 x,y 坐标
        x, y = point[0], point[1]
        plt.scatter(x, y, c='r', s=40)  # 绘制红色圆点
    
    plt.axis('off')
    plt.show()
```
